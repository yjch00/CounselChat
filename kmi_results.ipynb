# 생성한 KMI 데이터셋 리뷰

```python
import json

with open('result/sample/KMI_sample_v2_id101~130.json', 'r') as f:
    all_data = json.load(f)
    
print(len(all_data))
```

## 특정 id의 대화 print

```python
id_list = [101, 130]

for id_to_view in range(id_list[0], id_list[1]+1):
    for data in all_data:
        if data['id'] == id_to_view:
            print(f"ID: {data['id']}")
            print(f"Category: {data['category']}")
            print(f"Context: {data['context']}")
            print(f"Turns: {data['turns']}")
            print()
            print('Dialogue:')
            for u in data['dialogue']:
                print(f"{u['role']} [{u['label']}]: {u['utterance_ko']}")
            print()
            print('-'*100)
            print()
```

## 생성한 전체 데이터셋의 Label 비율 계산
- 각 대화를 여는 상담사의 OQ는 제외하고 count (이건 항상 고정이므로)

```python
from collections import Counter

total_turns = 0
therapist_labels = []
client_labels = []
therapist_label_order = ['Simple Reflection', 'Complex Reflection', 'Open Question', 'Closed Question', 'Affirm', 'Give Information', 'Advise', 'General']
client_label_order = ['Change Talk', None]

for data in all_data:
    total_turns += data['turns']
    for i, u in enumerate(data['dialogue']):
        if i != 0:
            if u['role'] == 'Therapist':
                therapist_labels.append(u['label'])

            elif u['role'] == 'Client':
                client_labels.append(u['label'])

therapist_counter_dict = dict(Counter(therapist_labels))
client_labels_dict = dict(Counter(client_labels))
therapist_total_count = sum(therapist_counter_dict.values())
client_total_count = sum(client_labels_dict.values())

print(f'Dialogue 개수: {len(all_data)}')
print(f'평균 턴 수: {round(total_turns/len(all_data), 2)}')
print()
print('[Therpist Labels]')
for label in therapist_label_order:
    count = therapist_counter_dict.get(label, 0)
    ratio = count / therapist_total_count
    print(f"{label}: {count} ({ratio:.2f})")
print()
print('[Client Labels]')
for label in client_label_order:
    count = client_labels_dict.get(label, 0)
    ratio = count / client_total_count
    print(f"{label}: {count} ({ratio:.2f})")
```
